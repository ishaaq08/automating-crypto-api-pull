```python
# This code has been copied from the Crypto API Documentation
    
from requests import Request, Session
from requests.exceptions import ConnectionError, Timeout, TooManyRedirects
import json

url = 'https://pro-api.coinmarketcap.com/v1/cryptocurrency/listings/latest'

parameters = {
  'start':'1',
  'limit':'15',
  'convert':'USD'
}
headers = {
  'Accepts': 'application/json',
  'X-CMC_PRO_API_KEY': 'e1f778a6-1860-4ebf-89ad-5f7b609e75da',
}

session = Session()
session.headers.update(headers)

try:
  response = session.get(url, params=parameters)
  data = json.loads(response.text)
except (ConnectionError, Timeout, TooManyRedirects) as e:
  print(e)
```

```python
import pandas as pd

pd.set_option("display.max_columns", None)
pd.set_option("display.max_rows", None)
```

```python
# JSON normalize - organise JSON data into a dataframe

df = pd.json_normalize(data["data"])
```

```python
df["timestamp"] = pd.to_datetime("now")
```

```python
# Automated process that will append data to the dataframe 

def api_runner():
    global df
    url = 'https://pro-api.coinmarketcap.com/v1/cryptocurrency/listings/latest'

    parameters = {
      'start':'1',
      'limit':'15',
      'convert':'USD'
    }
    headers = {
      'Accepts': 'application/json',
      'X-CMC_PRO_API_KEY': 'e1f778a6-1860-4ebf-89ad-5f7b609e75da',
    }

    session = Session()
    session.headers.update(headers)

    try:
      response = session.get(url, params=parameters)
      data = json.loads(response.text)
    except (ConnectionError, Timeout, TooManyRedirects) as e:
      print(e)
    
    # Format the data
    
    df2 = pd.json_normalize(data["data"])
    df2["timestamp"] = pd.to_datetime("now")
    df = pd.concat([df, df2])
```

```python
# Give us the ability to track the time so we can call the function in certain intervals

import os 
from time import time
from time import sleep

for i in range(333): 
    api_runner()
    print("API Runner completed successfully")
    sleep(60)
exit()
```
